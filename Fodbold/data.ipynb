```python
events = sb.events(match_id=3895302)
events
```

```python
events_passes = events[events['pass_recipient_id'].notnull()]
events_passes_filtered = events_passes[["location","pass_end_location","pass_goal_assist"]]
```

```python
events_passes_filtered["start_x"] = events_passes_filtered["location"].str[0]
events_passes_filtered["start_y"] = events_passes_filtered["location"].str[1]
events_passes_filtered["end_x"] = events_passes_filtered["pass_end_location"].str[0]
events_passes_filtered["end_y"] = events_passes_filtered["pass_end_location"].str[1]
```

```python
x_bins = np.linspace(0, events_passes_filtered["start_x"].max(), 7)
x_labels = [0, 1, 2, 3, 4, 5]

y_bins = np.linspace(0, events_passes_filtered["start_y"].max(), 4)
y_labels = [0, 1, 2]

events_passes_filtered["x_box_start"] = pd.cut(events_passes_filtered['start_x'], bins=x_bins, labels=x_labels, include_lowest=True)
events_passes_filtered["y_box_start"] = pd.cut(events_passes_filtered['start_y'], bins=y_bins, labels=y_labels, include_lowest=True)
events_passes_filtered["x_box_end"] = pd.cut(events_passes_filtered['end_x'], bins=x_bins, labels=x_labels, include_lowest=True)
events_passes_filtered["y_box_end"] = pd.cut(events_passes_filtered['end_y'], bins=y_bins, labels=y_labels, include_lowest=True)
```

```python

edge_df = pd.DataFrame() 
edge_df['box_start'] = list(zip(events_passes_filtered["x_box_start"], events_passes_filtered["y_box_start"]))
edge_df['box_end'] = list(zip(events_passes_filtered["x_box_end"], events_passes_filtered["y_box_end"]))
```

```python
edge_weights = edge_df.groupby(['box_start', 'box_end'], as_index=False).size()
```
