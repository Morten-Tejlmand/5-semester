```python
events_passes = events[events['pass_recipient_id'].notnull()]
events_passes_filtered = events_passes[["location","pass_end_location","pass_goal_assist"]]
```

```python
events_passes_filtered["start_x"] = events_passes_filtered["location"].str[0]
events_passes_filtered["start_y"] = events_passes_filtered["location"].str[1]
events_passes_filtered["end_x"] = events_passes_filtered["pass_end_location"].str[0]
events_passes_filtered["end_y"] = events_passes_filtered["pass_end_location"].str[1]
```

```python
x_len_start = events_passes_filtered["start_x"].max() // 6
y_len_start = events_passes_filtered["start_y"].max() // 3

x_len_end = events_passes_filtered["end_x"].max() // 6
y_len_end = events_passes_filtered["end_y"].max() // 3
```

```python
import networkx as nx
```

```python
events_passes_filtered['start_node_x'] = events_passes_filtered['start_x'] // 10
events_passes_filtered['start_node_y'] = events_passes_filtered['start_y'] // 10
events_passes_filtered['end_node_x'] = events_passes_filtered['end_x'] // 10
events_passes_filtered['end_node_y'] = events_passes_filtered['end_y'] // 10
```

```python
events_passes_filtered["start_node"] = events_passes_filtered["start_node_x"] + events_passes_filtered["start_node_y"] / 10
events_passes_filtered["end_node"] = events_passes_filtered["end_node_x"] + events_passes_filtered["start_node_y"] / 10
events_passes_filtered = events_passes_filtered.drop(['start_x', 'start_y',"end_x","end_y",'location', 'pass_end_location',"start_node_x","start_node_y","end_node_y","end_node_x"], axis=1)
```

```python
events_passes_filtered
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Example DataFrame
data = events_passes_filtered
df = pd.DataFrame(data)

# Create an undirected graph
G = nx.Graph()

# Calculate the frequency of each combination (start_node, end_node)
edge_weights = df.groupby(['start_node', 'end_node']).size().reset_index(name='weight')

# Add edges to the graph with weights
for index, row in edge_weights.iterrows():
    G.add_edge(row['start_node'], row['end_node'], weight=row['weight'])

# Get positions for the nodes
pos = nx.spring_layout(G)

# Get edge weights
edges = G.edges(data=True)
weights = [edge[2]['weight'] for edge in edges]

# Normalize edge weights for coloring
norm = plt.Normalize(min(weights), max(weights))
colors = [plt.cm.viridis(norm(w)) for w in weights]

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# Draw edges with colors based on weight
nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color=colors, width=2)

# Draw labels for nodes
nx.draw_networkx_labels(G, pos, font_size=12, font_family="sans-serif")

# Draw edge labels (weights)
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Create a scalar mappable for the colorbar
sm = plt.cm.ScalarMappable(cmap=plt.cm.viridis, norm=norm)
sm.set_array([])  # Empty array for the colorbar

# Add the colorbar
plt.colorbar(sm, label="Edge Weight")

# Show the plot
plt.title("Undirected Graph with Colored Edges Based on Weight")
plt.show()

```
